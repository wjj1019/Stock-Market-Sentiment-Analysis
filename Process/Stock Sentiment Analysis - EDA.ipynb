```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
from wordcloud import WordCloud
```

```python
stock = pd.read_csv('stock_data.csv')
```

Imbalanced Datasets

```python
sns.heatmap(stock.isnull())
```

```python
stock.describe()
```

```python
positive = stock[stock['Sentiment'] == 1]['Text']
negative = stock[stock['Sentiment'] == -1]['Text']
```

```python
#Positive Sentiments Texts
common_words = ''
for i in positive.values:
    i = str(i)
    tokens = i.split()
    common_words += " ".join(tokens)+" "
    
wc = WordCloud().generate(common_words)
plt.figure(figsize=(12,10))
plt.imshow(wc, interpolation = 'bilinear')
plt.title('Positive Sentiment Texts')
plt.axis('off')
plt.show()
```

```python
#Negative Sentiments Text
common_words = ''
for i in negative.values:
    i = str(i)
    tokens = i.split()
    common_words += " ".join(tokens)+" "
    
wc = WordCloud().generate(common_words)
plt.figure(figsize=(12,10))
plt.imshow(wc, interpolation = 'bilinear')
plt.title('Negative Sentiment Texts')
plt.axis('off')
plt.show()
```
